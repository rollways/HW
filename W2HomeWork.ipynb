### 1.抓出銷售資料裡面，2020-11-01~2020-11-30期間的銷售資料

### 2.刪除UnitPrice<0 或是UnitPrice > 3000的銷售資料​

```python
data2 = data1[ (data1.UnitPrice>0) & (data1.UnitPrice<3000) ]
data2
```
